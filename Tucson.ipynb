```python
import os
import shutil
```

```python
angle_list = os.listdir(r'D:\jupyter\연구 COLAB\YOLO\kcar\Hyundai\Hyundai\Hyundai\G80 스포츠\2017\10시대')
angle_list
```

```python
def read_all_file(path):
  output = os.listdir(path)
  file_list = []

  for i in output:
    for j in angle_list:
      final_list = os.listdir(path + '/' + i + '/' + j)
      for k in final_list :
        file_list.append(path + '/' + i + '/' + j + '/' + k)

  return file_list
```

```python
def copy_all_file(file_list, new_path):
  for src_path in file_list:
    file = src_path.split('/')[-1]
    shutil.copyfile(src_path, new_path+'/'+file)
    print('{} 완료'.format(file))
```

```python
j = {'G80 스포츠':"2017",'G90':"2018",'제네시스 G70':"2017",'제네시스 G80':"2018",'제네시스 G380':"2014"}

for a, b in j.items():
    src_path = 'D:/jupyter/연구 COLAB/YOLO/kcar/Hyundai/Hyundai/Hyundai/'+a+'/'+b
    new_path = 'D:/jupyter/연구 COLAB/YOLO/kcar/Hyundai/제네시스'


    file_list = read_all_file(src_path)
    # print(file_list)
    copy_all_file(file_list, new_path)


```
